```python

import numpy as np
import matplotlib as plt 
# Reload the saved file
data = np.load("predictions.npz")
pred_expr_dev = data["pred_expr_dev"]
Y_test_dev = data["Y_test_dev"]
pred_expr_hk = data["pred_expr_hk"]
Y_test_hk = data["Y_test_hk"]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to create 2D histogram density plots
def plot_density(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(6, 6))
    plt.hist2d(x, y, bins=50, range=[[0, 2.1], [0, 2.1]], cmap="Blues", density=True)
    plt.colorbar(label="Density")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.axline((0, 0), slope=1, color='red', linestyle="--", label="Perfect Correlation")  # 1:1 line
    plt.legend()
    plt.show()

# Plot for Dev_log2_enrichment
plot_density(Y_test_dev, pred_expr_dev, 
             "Density Plot: Predicted vs. Actual CMP(FLI1)", 
             "Actual FLI1_log10_enrichment", 
             "Predicted FLI1_log10_enrichment")

# Plot for Hk_log2_enrichment
plot_density(Y_test_hk, pred_expr_hk, 
             "Density Plot: Predicted vs. Actual CMP(PU1)", 
             "Actual PU1_log10_enrichment", 
             "Predicted PU1_log10_enrichment")
```
